```python
import pandas as pd
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from glob import glob
```

```python
def load_results(parameter_to_study, value_type, base_dir="."):
    res = pd.DataFrame()
    for idir in sorted(glob(os.path.join(base_dir, parameter_to_study, "*"))):
        ifn = os.path.join(idir, "topk_mm.tsv")
        
        if not os.path.exists(ifn):
            continue
        
        _res = pd.read_csv(ifn, sep="\t")
        _res[parameter_to_study] = value_type(idir.split(os.sep)[-1])
        res = pd.concat((res, _res), axis=0)

    res["top_k_acc"] = res["n_top_k"] / res["sequence_length"] * 100
    
    return res
```

## C

```python
parameter_to_study = "C"
res = load_results(parameter_to_study, float, "logs/parameter_study__ECFP__binary__all__metfrag__norm")
```

```python
sns.catplot(data=res[res["k"] <= 20], x="k", y="top_k_acc", col=parameter_to_study, hue="method", kind="point", legend_out=False, col_wrap=3)
sns.catplot(data=res[res["k"].isin([1, 5, 10, 20])], x=parameter_to_study, y="top_k_acc", col="k", hue="method", kind="point", sharey=False, legend_out=False)
sns.catplot(data=res[(res["k"] == 1) & (res["C"] == 16)], x="dataset", y="top_k_acc", col="k", hue="method", kind="box", sharey=False, legend_out=False, aspect=3)
```
